```python
import twint
import nest_asyncio
```

```python
nest_asyncio.apply()
c = twint.Config()
c.Search = "lockdown" 
c.Pandas = True
c.Limit = 200
c.Pandas_clean = True
#c.Count = 1000
c.Geo = "28.5934,77.2223,2000km"# Run
```

```python
dates = ['2020-03-25',
         '2020-03-26',
         '2020-03-27',
         '2020-03-28',
         '2020-03-29',
         '2020-03-30',
         '2020-03-31',
         '2020-04-01',
         '2020-04-02',
         '2020-04-03',
         '2020-04-04',
         '2020-04-05',
         '2020-04-06',
         '2020-04-07',
         '2020-04-08',
         '2020-04-09',
         '2020-04-10',
         '2020-04-11',
         '2020-04-12',
         '2020-04-13',
         '2020-04-14',
         '2020-04-15',
         '2020-04-16',
         '2020-04-17',
         '2020-04-18',
        ]

dates_ = ['2020-04-18',
          '2020-04-19',
          '2020-04-20',
          '2020-04-21',
          '2020-04-22',
          '2020-04-23',
          '2020-04-24',
          '2020-04-25',
          '2020-04-26',
          '2020-04-27',
          '2020-04-28',
          '2020-04-29',
          '2020-04-30',
          '2020-05-01',
          '2020-05-02',
          '2020-05-03',
         ]
```

```python
df = pd.DataFrame()
for i in range(0,len(dates)-1):
    c.Since = dates[i]
    c.Until = dates[i+1]
    c.Limit = 200
    c.Lang = 'en'
    c.Hide_output = True
    twint.run.Search(c)
    df_temp = pd.DataFrame(twint.storage.panda.Tweets_df)
    columns = ['date','tweet', 'hashtags',]
    df_temp = df_temp[columns]
    df = df.append(df_temp , ignore_index = True)
    print("\n Current: ",dates[i], " ", df.shape)
```
